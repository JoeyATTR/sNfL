# Auswertung (Berufspraktikum_Joel)

```python
import pandas as pd
import numpy as np        

path = "joel_export_mitsnfl_verblindet_anon1.xlsx"
df = pd.read_excel(path)

print(df.shape)        
print(df.head())      

COLS = {
    "PatientID_PSD": "PatientID", "AmyloidosisType": "AmyType", "baseline_age": "Age", "baseline_nis": "NisScore", "baseline_dpd": "Perugini", "sex": "Sex", 
    "Neurofilament-Light (NF-L) Se": "sNFL",          
    "GFR_ml_per_min": "eGFR",
    "nt_pro_bnp_percentage": "NTproBNP",              
    "carpal_tunnel_near_baseline": "CTS",
    "spinal_canal_near_baseline": "SCS",                   
    "ortho_joint_replacement": "JointReplacement",
    "onco_any_comorbidity": "Onco",
    "toxic_alcohol_abuse": "AlcoholAbuse",
    "diabetis_mellitus": "Diabetis",
    "chronic_kidney_disease": "KidneyDisease",
    "history_of_chemotherapy": "HisOfChemo",
    "exposure_to_toxic_substances": "ToxicExposure",
}

cols_present = {k:v for k,v in COLS.items() if k in df.columns}    
df = df.rename(columns=cols_present)

def normalize_to_bool01(s):                                          
    TRUE_VALUES = {"yes","ja","true","wahr","y","j","1"}
    FALSE_VALUES = {"no","nein","false","falsch","n","0"}            
    def one(x):
        if pd.isna(x): return 0     
        if isinstance(x, bool): return 1 if x else 0                 
        if isinstance(x, (int, float)): return 1 if x > 0 else 0
        t = str(x).strip().lower()                  
        if t in TRUE_VALUES: return 1
        if t in FALSE_VALUES: return 0
        return 0
    return s.apply(one) 

COMORB_VARS = ["Onco","AlcoholAbuse","Diabetis", "JointReplacement","KidneyDisease","HisOfChemo","ToxicExposure"]
present_cols = [c for c in COMORB_VARS if c in df.columns]
for c in present_cols:
    df[c] = normalize_to_bool01(df[c])
  
any_comorb = df[present_cols].sum(axis=1) > 0                          
print("Fälle mit ≥1 Komorbidität (werden ausgeschlossen):", int(any_comorb.sum()))   
print("Gesamtfälle vorher:", len(df))    


before = len(df)
df["AmyType"] = (df["AmyType"] != "ATTR-Amyloidose, Wild-Typ").astype(int)
df = df.loc[df["AmyType"] == 0].copy()
after = len(df)
print(f"Nach Auswahl nur Wild-Typ: {after} von {before} Fällen übrig")

# --- Perugini II/II-III/III = 0, alle anderen = 1 ---
before = len(df)
df["Perugini"] = df["Perugini"].astype(str).str.strip()
df["Perugini"] = df["Perugini"].apply(lambda x: 0 if x in ["Perugini II", "Perugini II-III", "Perugini III"] else 1)
df = df.loc[df["Perugini"] == 0].copy()
after = len(df)
print(f"Nach Auswahl Perugini II–III: {after} von {before} Fällen übrig")

# --- NIS in binär: >=20 = 1 ---
df["NisScore"] = (df["NisScore"] >= 20).astype(int)            

# --- Sex in binär: m=1, w=0 (alles bleibt erhalten) ---
df["Sex"] = df["Sex"].astype(str).str.strip().str.lower()
df["Sex"] = df["Sex"].map({"m":1, "male":1, "w":0, "female":0}).fillna(df["Sex"])

before = len(df)
df = df.loc[~any_comorb].copy()                           
after = len(df)
print(f"Ausgeschlossen: {before - after}  |  Verbleibend: {after}")

df = df.dropna(subset=["NisScore", "sNFL", "NTproBNP", "Age", "eGFR"], how="any")   

# --- bestimmte Spalten komplett entfernen ---
drop_cols = ["spinal_canal_stenosis_present", "manifestation", "carpal_tunnel_syndrome_left_right", 
             "autoimmune_rheumatological", "pnd_score"]
df = df.drop(columns=[c for c in drop_cols if c in df.columns])

df.head()
```

```python
!pip install scikit-learn
from sklearn.metrics import roc_curve, roc_auc_score


y = df["NisScore"]        # Das ist meine Zielvariable


x = df["sNFL"]           # Das ist mein Prädikator


auc = roc_auc_score(y, x)     # AUC berechnen 


fpr, tpr, thresholds = roc_curve(y, x)     # ROC-Kurve berechnen (liefert FPR, TPR und Schwellenwerte)


youden_index = (tpr - fpr).argmax()            # Youden-Index (maximiert Sens + Spez - 1)
cutoff = thresholds[youden_index]
sens = tpr[youden_index]
spez = 1 - fpr[youden_index]

print(f"AUC sNFL = {auc:.3f}")
print(f"Optimaler Cut-off nach Youden: {cutoff:.2f}")
print(f"Sensitivität = {sens:.2f}")
print(f"Spezifität = {spez:.2f}")
```

```python
import statsmodels.api as sm
import pandas as pd

# --- Funktion für schnelles Modell ---
def run_logit(df, y_var, x_vars):
    """
    df     = df
    y_var  = NisScore
    x_vars = sNFL
    """
    # 1) Y und X auswählen
    Y = df[y_var]
    X = df[x_vars]

    # 2) Alles numerisch machen und NaN raus
    data = pd.concat([Y, X], axis=1).apply(pd.to_numeric, errors="coerce").dropna()
    Y = df[y_var]
    X = sm.add_constant(data[x_vars])   # Konstante hinzufügen

    # 3) Modell fitten
    model = sm.Logit(Y, X).fit(disp=0)
    print(model.summary())
    return model

# --- Beispielaufrufe ---
# 1) Nur sNFL → NIS
run_logit(df, "NisScore", ["sNFL"])

# 2) Vergleich sNFL + NTproBNP
run_logit(df, "NisScore", ["sNFL", "NTproBNP"])
```

# sNFL und NTproBNP müssen Log-Transformiert werden (nutzbare Skalierung notwendig)

```python
import statsmodels.api as sm
import pandas as pd
import numpy as np

# --- Funktion für schnelles Modell mit log-Transform ---
def run_logit(df, y_var, x_vars, log_vars=True):
    """
    df     = Datensatz
    y_var  = Zielvariable (z.B. NisScore)
    x_vars = Liste mit Prädiktoren (z.B. ["sNFL"])
    log_vars = True → alle Prädiktoren werden log-transformiert
    """
    # 1) Y und X auswählen
    Y = df[y_var]
    X = df[x_vars]

    # 2) Alles numerisch machen und NaN raus
    data = pd.concat([Y, X], axis=1).apply(pd.to_numeric, errors="coerce").dropna()
    Y = data[y_var]
    X = data[x_vars]

    # 3) Log-Transformation (optional)
    if log_vars:
        X = np.log(X + 1)   # +1 verhindert log(0)

    # 4) Konstante hinzufügen
    X = sm.add_constant(X)

    # 5) Modell fitten
    model = sm.Logit(Y, X).fit(disp=0)
    print(model.summary())
    return model


# --- Beispiele ---
# Nur sNFL → NIS
run_logit(df, "NisScore", ["sNFL"])

# Vergleich sNFL + NTproBNP
run_logit(df, "NisScore", ["sNFL", "NTproBNP"])
```

```python
import numpy as np, pandas as pd, statsmodels.api as sm

def logit_or_per_sd(df, y_var="NisScore",
                    x_vars=("sNFL","Age","eGFR","SCS","JointReplacement"),
                    drop_corr_thresh=0.999):
    # --- Daten auswählen & numerisch machen ---
    cols = [c for c in x_vars if c in df.columns]
    data = df[[y_var]+cols].copy().replace({True:1, False:0})
    # Sex o.ä. ggf. schon 0/1 im df
    data = data.apply(pd.to_numeric, errors="coerce").dropna()

    # --- pro 1 SD standardisieren (nur stetige) ---
    for c in ("sNFL","Age","eGFR"):
        if c in data.columns:
            sd = data[c].std(ddof=0)
            if sd and sd > 0: data[c] = (data[c]-data[c].mean())/sd

    Y = data[y_var].astype(int)
    X = data[cols].astype(float)

    # --- Diagnostik & Bereinigung ---
    removed = {}

    # 0-Varianz
    zero_var = [c for c in X.columns if X[c].var() == 0]
    if zero_var:
        removed["zero_var"] = zero_var
        X = X.drop(columns=zero_var)

    # Duplikate
    dup = list(X.columns[X.T.duplicated()])
    if dup:
        removed["duplicates"] = dup
        X = X.drop(columns=dup)

    # Nahezu perfekte Korrelation
    if X.shape[1] > 1:
        corr = X.corr().abs()
        upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
        drop_high = [c for c in upper.columns if any(upper[c] > drop_corr_thresh)]
        if drop_high:
            removed["high_corr"] = drop_high
            X = X.drop(columns=drop_high)

    # Wenn jetzt leer, abbrechen
    if X.shape[1] == 0:
        raise ValueError("Alle Prädiktoren sind konstant/dupliziert/hoch korreliert. Wähle andere Variablen.")

    # Konstante zuletzt hinzufügen
    Xc = sm.add_constant(X)

    # --- Fit: zuerst normales Logit, sonst Regularisierung ---
    try:
        m = sm.Logit(Y, Xc).fit(disp=0)
        has_se = True
    except np.linalg.LinAlgError:
        m = sm.Logit(Y, Xc).fit_regularized(method="l1", maxiter=2000)
        has_se = False  # keine p/CI bei fit_regularized

    # --- Ausgabe (OR pro 1 SD für standardisierte Variablen) ---
    names = list(getattr(m, "params").index)
    if has_se:
        OR = np.exp(m.params); CI = np.exp(m.conf_int())
        res = pd.DataFrame({
            "Variable": names,
            "OR": OR.values,
            "CI_low": CI[0].values,
            "CI_high": CI[1].values,
            "p": m.pvalues.values
        }).round(3)
    else:
        OR = np.exp(pd.Series(m.params, index=names))
        res = pd.DataFrame({"Variable": names, "OR": OR.values}).round(3)

    if removed:
        print("Bereinigt:", removed)
    print(res)
    return m, res

# Aufruf mit deinen Kovariaten (Age, eGFR, SCS, JointReplacement) und sNFL:
model, table = logit_or_per_sd(df, y_var="NisScore",
                               x_vars=("sNFL","Age","eGFR","SCS","JointReplacement"))
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

def logit_with_youden(df, y_var="NisScore", cutoff=45.2):      # Youden Cut-Off als binär im Modell eingesetzt und die passende OR dafür berechnet 
    """
    df      = dein DataFrame
    y_var   = Zielvariable (hier: NisScore >= 20)             
    cutoff  = Youden-Cut-off für sNFL
    """
                                                               # Neue binäre Variable für sNFL
    df = df.copy()
    df["sNFL_high"] = (df["sNFL"] >= cutoff).astype(int)

                                                               # Prädiktoren auswählen (inkl. binär sNFL_high)
    x_vars = ["sNFL_high", "Age", "eGFR", "SCS"]
    data = df[[y_var] + x_vars].replace({True:1, False:0})
    data = data.apply(pd.to_numeric, errors="coerce").dropna()   # Alle nicht numerischen Werte oder fehlende Werte wie NaN werden automatisch rausgefiltert 

    # Y und X
    Y = data[y_var].astype(int)
    X = sm.add_constant(data[x_vars])                        #In einem multiplen Regressionsmodell stellt die Konstante den Wert dar, der für die abhängige Variable vorhergesagt würde, wenn alle unabhängigen Variablen gleichzeitig gleich Null wären

    # Logit fitten                                            # kann zur Linearisierung von sigmoiden Kurven verwendet werden 
    model = sm.Logit(Y, X).fit(disp=0)

    # OR + CI berechnen                                       # (OR) -> Der Wert zeigt, wie viel mal höheres/niedrigeres Risiko (oder Chance) in der exponierten Gruppe im Vergleich zur nicht exponierten Gruppe vorliegt. Ein OR > 1 bedeutet ein höheres Risiko, ein OR < 1 ein geringeres Risiko. 
    OR = np.exp(model.params)                                 # Das CI gibt einen Wertebereich an, in dem der wahre Wert des Odds Ratio mit einer bestimmten Wahrscheinlichkeit (z.B. 95%) liegt. 
    CI = np.exp(model.conf_int())
    res = pd.DataFrame({
        "Variable": OR.index,
        "OR": OR.values,
        "CI_low": CI[0].values,             
        "CI_high": CI[1].values,
        "p": model.pvalues.values
    }).round(3)

    print(res)
    return model, res

model_youden, table_youden = logit_with_youden(df, y_var="NisScore", cutoff=45.2)
```

# DeLong-Test: sNFL vs. NTproBNP

```python
import numpy as np
from sklearn.metrics import roc_auc_score, roc_curve
from scipy import stats

# ---- Hilfsfunktionen (DeLong) ----
def compute_midrank(x):
    """Midranks für DeLong-Berechnung"""
    J = np.argsort(x)
    Z = x[J]
    N = len(x)
    T = np.zeros(N, dtype=float)
    i = 0
    while i < N:
        j = i
        while j < N and Z[j] == Z[i]:
            j += 1
        T[i:j] = 0.5*(i + j - 1)
        i = j
    T2 = np.empty(N, dtype=float)
    T2[J] = T + 1
    return T2

def fastDeLong(y_true, y_scores):
    """Varianz-berechnetes AUC nach DeLong"""
    y_true = np.array(y_true)
    assert set(y_true) <= {0,1}
    pos = y_scores[y_true==1]
    neg = y_scores[y_true==0]
    m, n = len(pos), len(neg)

    pos_ranks = compute_midrank(np.concatenate([pos, neg]))[:m]
    auc = (pos_ranks - (m+1)/2).sum() / (m*n)
    v01 = (pos_ranks - (m+1)/2) / n
    v10 = ((m*n) - (pos_ranks - (m+1)/2)) / m
    sx = np.var(v01, ddof=1)
    sy = np.var(v10, ddof=1)
    se_auc = np.sqrt(sx/m + sy/n)
    return auc, se_auc

def delong_test(y_true, scores1, scores2):
    """Vergleich zweier AUCs mit DeLong"""
    auc1, se1 = fastDeLong(y_true, scores1)
    auc2, se2 = fastDeLong(y_true, scores2)
    diff = auc1 - auc2
    se_diff = np.sqrt(se1**2 + se2**2)
    z = diff / se_diff
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return auc1, auc2, diff, z, p

# ---- Beispiel ----
y = df["NisScore"].astype(int).values
s1 = df["sNFL"].values
s2 = df["NTproBNP"].values

auc1, auc2, diff, z, p = delong_test(y, s1, s2)
print(f"AUC sNFL     = {auc1:.3f}")
print(f"AUC NTproBNP = {auc2:.3f}")
print(f"Differenz    = {diff:.3f}")
print(f"Z-Statistik  = {z:.3f}")
print(f"p-Wert       = {p:.4f}")
```

```python
# ===== Vergleich: Bootstrap-Optimismus für 3 Modelle (AUC) =====
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression   # logistische Regression
from sklearn.metrics import roc_auc_score              # AUC-Berechnung
from sklearn.utils import resample                     # Bootstrap-Resampling
import matplotlib.pyplot as plt                        # für Plot

# --- Hilfsfunktion: berechnet apparent AUC, Optimismus & korrigierte AUC ---
def auc_bootstrap_optimism(X, y, n_boot=500, seed=42):
    """
    X: DataFrame mit Prädiktoren (nur numerisch, keine NaNs)
    y: Series (0/1)
    n_boot: Anzahl Bootstrap-Resamples
    """
    # 1) Modell auf Originaldaten trainieren
    m0 = LogisticRegression(max_iter=5000, solver="liblinear").fit(X, y)
    p0 = m0.predict_proba(X)[:,1]              # vorhergesagte Wahrscheinlichkeiten
    auc_app = roc_auc_score(y, p0)             # apparent AUC (optimistisch)

    # 2) Bootstrap-Schleife starten
    app, test = [], []                         # Listen für apparent/test AUC
    rng = np.random.default_rng(seed)          # Zufallszahlengenerator für Reproduzierbarkeit
    for _ in range(n_boot):                    # wiederhole n_boot-mal
        idx = resample(X.index, replace=True,  # Stichprobe mit Zurücklegen
                       random_state=int(rng.integers(1e9)))
        Xb, yb = X.loc[idx], y.loc[idx]        # Bootstrap-Datensatz
        m = LogisticRegression(max_iter=5000, solver="liblinear").fit(Xb, yb)
        app.append(roc_auc_score(yb, m.predict_proba(Xb)[:,1]))  # AUC im Bootstrap-Sample
        test.append(roc_auc_score(y,  m.predict_proba(X) [:,1])) # AUC auf Originaldaten

    # 3) Optimismus = mittlere Differenz zwischen apparent und test
    optimism = float(np.mean(np.array(app) - np.array(test)))
    auc_corr  = auc_app - optimism             # bias-korrigierte AUC
    return auc_app, optimism, auc_corr

# --- Daten vorbereiten (numerisch + ohne NaNs) ---
def clean_xy(df, y_col, x_cols):
    Z = df[[y_col] + x_cols].apply(pd.to_numeric, errors="coerce").dropna()  # nur numerisch & keine NaNs
    y = Z[y_col].astype(int)                    # Zielvariable (0/1)
    X = Z.drop(columns=y_col).astype(float)     # Prädiktoren
    return X, y

# 1) Modell mit sNFL allein
X1, y1 = clean_xy(df, "NisScore", ["sNFL"])
a1, o1, c1 = auc_bootstrap_optimism(X1, y1)

# 2) Modell mit NT-proBNP allein
X2, y2 = clean_xy(df, "NisScore", ["NTproBNP"])
a2, o2, c2 = auc_bootstrap_optimism(X2, y2)

# 3) Multivariates Modell (sNFL + Age + eGFR + SCS)
X3, y3 = clean_xy(df, "NisScore", ["sNFL","Age","eGFR","SCS"])
X3 = X3.loc[:, X3.var() > 0]                   # Spalten ohne Varianz entfernen
a3, o3, c3 = auc_bootstrap_optimism(X3, y3)

# --- Ergebnisse in Tabelle packen ---
tab = pd.DataFrame({
    "Modell": ["sNFL", "NTproBNP", "sNFL+Age+eGFR+SCS"],  # Name der Modelle
    "AUC_apparent": [a1, a2, a3],                        # apparent AUC
    "Optimismus":   [o1, o2, o3],                        # mittlere Überschätzung
    "AUC_korrigiert":[c1, c2, c3],                       # bias-korrigierte AUC
}).round(3)

print(tab)  # Tabelle ausgeben

# --- Balkenplot: apparent vs. korrigiert ---
labels = tab["Modell"].values          # Modellnamen
x = np.arange(len(labels))             # x-Achse
w = 0.35                               # Balkenbreite

plt.figure(figsize=(7.2,4.2))
plt.bar(x - w/2, tab["AUC_apparent"], width=w, label="apparent")     # apparent Balken
plt.bar(x + w/2, tab["AUC_korrigiert"], width=w, label="bias-korrigiert") # korrigierte Balken
plt.axhline(0.5, linestyle="--")        # Linie für Zufall (AUC=0.5)
plt.xticks(x, labels, rotation=0)       # Beschriftung
plt.ylim(0.4, 0.85)                     # Y-Achse begrenzen
plt.ylabel("AUC")
plt.title("AUC-Vergleich mit Bootstrap-Optimismus")
plt.legend()
plt.tight_layout()
plt.show()
```

## Blaue Balken -> apparent AUC (ohne Korrektur)
## Orange Balken -> bias-korrigierte AUC (realistisch) 
## Gestrichelte Linie bei 0.5: Zufallsniveau
### -> NTproBNP allein taugt nicht als Prädiktor
### -> sNFL hat eine moderate Aussagekraft, die nach Bias-Korrektur stabil bleibt 
### -> kombiniertes Modell sieht gut aus, ist aber überoptmistisch -> nach Korrektur nur wenig besser als sNFL allein 

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score


def cv_auc(df, y_col, x_cols, k=5, seed=42):  #Definiert eine Funktion cv_auc -> berechnet CV-AUC für beliebige Prädiktoren
  
    Z = df[[y_col] + x_cols].apply(pd.to_numeric, errors="coerce").dropna()  # nimmt nur relevante Spalten aus df, wandelt in numerische Werte um und entfernt NaN
    y = Z[y_col].astype(int).values                   #Zielvariable (NisScore 0/1)
    X = Z.drop(columns=y_col).astype(float).values    # alle Prädiktoren als Matrix

   
    skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=seed) #Erstellt eine Cross-Validation mit k Folds; shuffle=TRUE -> zufällige Verteilung, aber random_state=seed sorgt für Reproduzierbarkeit

    aucs = []   # Leere Liste für AUC-Werte       #Daten werden in Trainings- und Testssets aufgeteilt
    for train, test in skf.split(X, y):
        m = LogisticRegression(max_iter=5000, solver="liblinear").fit(X[train], y[train])  # Modell wird auf dem Trainingsdatensatz fitten
        p = m.predict_proba(X[test])[:,1]    # Wahrscheinlichkeiten für Klasse 1 (NIS ≥ 20)
        aucs.append(roc_auc_score(y[test], p))   # berechnet die AUC für die Testdaten
    return np.array(aucs)   # -> Gibt die Liste der AUC-Werte (einen pro Fold) als numpy-Array zurück


aucs_s = cv_auc(df, "NisScore", ["sNFL"])
print(f"sNFL 5-fold CV-AUC:  Mittel={aucs_s.mean():.3f}  SD={aucs_s.std(ddof=1):.3f}")  # Berechnet 5-fold CV für sNFL + Durchschnittliche AUC über alle 5 Folds + std(ddof=1) = Streuung -> zeigt Stabilität


aucs_m = cv_auc(df, "NisScore", ["sNFL","Age","eGFR","SCS"])      # Dasselbe für das multivariable Modell mit sNFL + Age + eGFR + SCS
print(f"sNFL+Age+eGFR+SCS 5-fold CV-AUC:  Mittel={aucs_m.mean():.3f}  SD={aucs_m.std(ddof=1):.3f}")
```

## SD = 0.097 -> relativ stabile Ergebnisse über die 5 Folds, aber etwas Schwankung 

# sNFL+AGE+eGFR+SCS:
## AUC = 0.659 -> leicht schlechter als sNFL allein
## SD = 0.129 -> höhere Unsicherheit, Modell schwankt stärker je nach Fold 
### Das Modell nur mit sNFL schneidet im Cross-Validation-Test sogar etwas besser ab als das Modell mit den zusätzlichn Kovariaten -> sNFL alleine stabilerer Prädiktor als die Kombination

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm


Z = df[["NisScore","sNFL","Age","eGFR","SCS"]].apply(pd.to_numeric, errors="coerce").dropna()  # nimmt nur die beötigten Spalten aus df + forced numerische Werte + entfernt Zeilen mit fehlenden Werten 


Z["log_sNFL"] = np.log(Z["sNFL"] + 1e-6)  # bildet den Logarithmus von sNFL (+1e-6 verhindert Probleme, falls ne 0 drin wäre) 


for v in ["log_sNFL","Age","eGFR"]:
    Z[v+"_z"] = (Z[v] - Z[v].mean()) / Z[v].std(ddof=0)  # z-Standardisierung: Mittelwert 0, SD 1


y = Z["NisScore"].astype(int) # ist das binäre Outcome
X = sm.add_constant(Z[["log_sNFL_z","Age_z","eGFR_z","SCS"]]) # enthält log_sNFL, Age_z, eGFR_z, SCS (binär) + eine Konstante, die add_constant automatisch einfügt 


model = sm.Logit(y, X).fit(disp=0)  # passt die logistische Regression an 
OR  = np.exp(model.params)          # Durch np.exp(model.params) bekommst du Odds Ratio
CI  = np.exp(model.conf_int())      # liefert 95%-Konfidenzintervalle (nach exp) sinds OR-Intervalle
out = pd.DataFrame({
    "Variable": OR.index,
    "OR_pro_1SD": OR.values,    
    "CI_low": CI[0].values,
    "CI_high": CI[1].values,
    "p": model.pvalues.values
}).round(3)

print(out)
```

## Age_z (OR ≈ 2.53, p ≈ 0.075) -> Pro 1 SD älter höhere Odds, aber nicht statistisch signifikant (möglicher Trend, jedoch zu kleine n) 

## eGFR_z (OR ≈ 0.83, p = 0.592) -> Pro 1 SD höhere eGFR weniger Odds, aber klar nicht signifikant -> kein Nachweis, dass Nierenfunktion in diesem Modell den NIS-Status erklärt (nach Adjustment)

## SCS (OR ≈ 1.29, p = 0.697) -> keine signifikante Assoziation
